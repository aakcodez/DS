Numpy and Pandas

Numpy Operations

```python
import numpy as np
```

```python
'''creating an array '''

a = np.array([1,2,3,4,5])
print("Array:", a)
```

```python
'''Reshaping array'''
b = np.array([1,2,3,4,5,6])
print("Reshaped Array:\n", b.reshape(2,3))
```

```python
'''Mathematical Operations'''

x = np.array([1,2,3,4])
y = np.array([3,4,5,6])

print("Addition:", x+y )
print("Mean:", np.mean(x))
print("Standard Devation:", np.std(x))
```

```python
'''Indexing  &  Slicing'''

arr = np.array([100,200,300,400,500])
print("Third element:", arr[2])
print("Slice:", arr[1:4])
```

```python
'''Random Number Generator'''

print("Random numbers:\n", 
      np.random.randint(3,10,5))
```

```python
'''Create a 2D static array'''

arr = np.array([[1,2,3],
               [4,5,6]])
print("2D array:\n", arr)
```

```python
print("Row-wise Std dev:",
      np.std(arr, axis=1))
```

```python
print("Column-wise Std dev:", 
      np.std(arr, axis=1))
```

```python
print("Overall Std dev:",
      np.std(arr))
```

```python
'''flatten array'''

flat = arr.flatten()
print("Flatten Array:", flat)
```

Pandas Operation

```python
import pandas as pd
```

```python
'''Creating DataFrame'''

a = {
    "Name": ["Anand", "Rahul", "Sneha"],
    "Marks": [85,90,95]
}
df = pd.DataFrame(a)
print(df)
```

```python
'''Viewing Data'''
print(df.head())
print(df.tail())
```

```python
'''Selecting Columns'''

print(df["Marks"])
print(df["Name"])
```

```python
'''Handling Missing Values'''
df.isnull()
```

```python
'''Grouping Data'''

print(df.groupby("Name")["Marks"].mean())
```

```python

a = {
    "Name": ["Anand", "Rahul", "Sneha"],
    "Marks": [85,90,95],
    "Age": [21,22,30]
}
df = pd.DataFrame(a)
print("First Data Frame:\n", df)
```

```python

a = {
    "Name": ["Anand", "Sneha"],
    "Marks": [85,90],
    "Age": [21,22]
}
df = pd.DataFrame(a)
print("Second Data Frame:\n", df)
```

```python
'''accessing the single row'''
print(df.loc[0])
```

```python
'''access column'''
print(df["Name"])
```

```python
print(df.iloc[0])
print(df.iloc[1,1])
```

```python
df_dropped=df.drop(0)
print("After Droping Row:\n",
      df_dropped)
```
