```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("mpg_raw.csv")
df
```

```python
plt.figure(figsize=(10,8))
df[["mpg", "acceleration", "weight", "displacement"]].hist()
plt.title("Histogram display using Raw Data")
plt.show()
```

```python
sns.scatterplot(data=df, x="weight", y="mpg")
plt.title("Scatter plot Plotting using Data")
plt.show()
```

```python
sns.countplot(data=df, x="cylinders")
plt.title("Count Plot representation")
plt.show()
```

```python
sns.pointplot(data=df, x="model_year", y="mpg")
plt.title("Point Plot display")
plt.show()
```
